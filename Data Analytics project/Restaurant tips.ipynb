"What is Skewness?"

Skewness is a statistical measure that tells us how much a distribution (data) is asymmetrical â€” that is, how much it leans or "skews" to the left or right, rather than being perfectly balanced like a normal (bell-shaped) distribution.

ðŸ“ˆ Types of Skewness:
Skewness = 0
â†’ Perfectly symmetric (e.g., normal distribution)

Positive Skew (Right Skew):
â†’ Tail is longer on the right side
â†’ Mean > Median
â†’ Example: Income data (most people earn less, a few earn a lot)

Negative Skew (Left Skew):
â†’ Tail is longer on the left side
â†’ Mean < Median
â†’ Example: Test scores where most students scored high, few scored low

```python
import matplotlib.pyplot as plt

male = df[df["sex"] == "Male"]
female = df[df["sex"] == "Female"]

plt.scatter(male["total_bill"], male["tip"], color="blue", label="Male")
plt.scatter(female["total_bill"], female["tip"], color="orange", label="Female")

plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Tip vs Total Bill by Gender")
plt.legend()
plt.show()
```
